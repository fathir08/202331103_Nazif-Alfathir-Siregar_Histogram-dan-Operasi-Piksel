```python
#202331103_Nazif Alfathir Siregar
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# Membaca Data dan Gambar

```python
#202331103_Nazif Alfathir Siregar
img = cv2.imread('Gambar Abu.png')
```

```python
#202331103_Nazif Alfathir Siregar
img.shape
```

```python
#202331103_Nazif Alfathir Siregar
[baris, kolom] = img.shape[:2]
```

# Convert BRG to RGB

```python
#202331103_Nazif Alfathir Siregar
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

# Menampilkan Histogram

```python
#202331103_Nazif Alfathir Siregar
fig, axs = plt.subplots(2,2,figsize=(15,5))

#cara1
axs[0,0].imshow(rgb)
axs[0,1].hist(rgb.ravel(), bins=256, range=[0,256])

#cara2
hist = cv2.calcHist([rgb],[0],None,[256],[0,256])
axs[1,0].imshow(rgb)
axs[1,1].plot(hist)

#menampilkan
plt.show()
```

# Operasi piksel untuk meningkatkan kecerahan

```python
#202331103_Nazif Alfathir Siregar
beta  = 40
citra_cerah = np.zeros((baris, kolom, 3))

for x in range(baris):
    for y in range(kolom):
        gyx = rgb[x,y] + beta
        citra_cerah[x,y] = gyx
citra_cerah = citra_cerah.astype(np.uint8)

fig, axs = plt.subplots(2,2,figsize=(15,5))
axs[0,0].imshow(rgb)
axs[0,1].hist(rgb.ravel(), bins=256, range=[0,256])
axs[1,0].imshow(citra_cerah)
axs[1,1].hist(citra_cerah.ravel(), bins=256, range=[0,256])

plt.show()
```

```python
#202331103_Nazif Alfathir Siregar
beta = 10
alpha = 1.2
citra_hasil = np.zeros((baris, kolom, 3))

for x in range(baris):
    for y in range(kolom):
        gyx = rgb[x,y] + beta
        citra_hasil[x,y] = gyx

citra_hasil = citra_hasil.astype(np.uint8)

fig, axs = plt.subplots(2,2,figsize=(15,5))
axs[0,0].imshow(rgb)
axs[0,1].hist(rgb.ravel(), bins=256, range=[0,256])
axs[1,0].imshow(citra_hasil)
axs[1,1].hist(citra_hasil.ravel(), bins=256, range=[0,256])

plt.show()
```

# Membalik Citra

```python
#202331103_Nazif Alfathir Siregar
citra_negatif = np.zeros((baris, kolom, 3))

for x in range(baris):
    for y in range(kolom):
        gyx = 225 - rgb[x,y]
        citra_negatif[x,y] = gyx

citra_negatif = citra_negatif.astype(np.uint8)

fig, axs = plt.subplots(2,2,figsize=(15,5))
axs[0,0].imshow(rgb)
axs[0,1].hist(rgb.ravel(), bins=256, range=[0,256])
axs[1,0].imshow(citra_negatif)
axs[1,1].hist(citra_negatif.ravel(), bins=256, range=[0,256])

plt.show()
```

# Meregangkan Kontras

```python
#202331103_Nazif Alfathir Siregar
alpha = 1.2
citra_kontras = np.zeros((baris, kolom, 3))

for x in range(baris):
    for y in range(kolom):
        gyx = rgb[x,y] * alpha
        citra_kontras[x,y] = gyx

citra_kontras = citra_kontras.astype(np.uint8)

fig, axs = plt.subplots(2,2,figsize=(15,5))
axs[0,0].imshow(rgb)
axs[0,1].hist(rgb.ravel(), bins=256, range=[0,256])
axs[1,0].imshow(citra_kontras)
axs[1,1].hist(citra_kontras.ravel(), bins=256, range=[0,256])

plt.show()
        
```
